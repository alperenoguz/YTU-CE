```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
headers={'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/113.0'}
```

```python
table=[]
linkbasi="https://www.hepsiburada.com"
head_table=[] # Alınan tablodaki başlıkların tutulduğu tablo
bas_links=[] # Verilerin çekileceği sayfaların linkinin tutulduğu dizi
for i in range(1,3):    # Kaçıncı sayfalardan veri alınacağının belirlenmesi. Çok fazla sayfadan veri almaya çalışınca bir süre sonra kernel ölebiliyor. 
                        # Her sayfada yaklaşık 20 tane bilgisayar var 
    bas_links.append('https://www.hepsiburada.com/laptop-notebook-dizustu-bilgisayarlar-c-98?sayfa='+str(i+1))
for bas_link in bas_links: #her sayfa için
    r= requests.get(bas_link, headers=headers) #Sayfaya giriş yapılması
    soup=BeautifulSoup(r.content,"lxml")
    
    st1=soup.find("div",attrs={"class":"productListContent-pXUkO4iHa51o_17CBibU"})
    st2=st1.find("ul",attrs={"class":"productListContent-frGrtf5XrVXRwJ05HUfU productListContent-rEYj2_8SETJUeqNhyzSm"})
    for links in st2.find_all("li",attrs={"class":"productListContent-zAP0Y5msy8OHn5z7T_K_"}): # Bu kısımda bilgisayarların linkleri alınıyor.
        values = []
        heads = []
        link=links.a.get("href")
        if "adservice" not in link:    #Alınan linkler bazen adservice şeklinde direk link olabildiğinden bu tarz bir kontrol uyguladım.
            link=linkbasi+link
        print(link) #Linkleri gidişatı izlemek için printliyorum.
        
        r2=requests.get(link,headers=headers) #Linke giriş yapılıyor. Bu linke giriş yapıldığında bilgisayarların kendi sayfasına girilmiş oluyor
        soup2=BeautifulSoup(r2.content,"lxml")
        
        pc_name=soup2.find("h1",attrs={"class":"product-name"}).text #Bilgisayarın isminin alınması
        pc_name=re.sub(r'[\n\t]', '',pc_name).strip() 
        values.append(pc_name) 
        
        st8=soup2.find("div",attrs={"class":"list-item-detail product-detail box-container"}) # Sayfadaki tabloyu bulma
        st9=st8.find("div",attrs={"id":"productTechSpecContainer"})
        st10=st9.find("table",attrs={"class":"data-list tech-spec"})
        
        for x in st10.find_all('tr'):     #Tablodaki Başlıklar ve değerlerin çekildiği yer
            for i in x.find_all('th'):
                heads.append(i.text)
            for k in x.find_all('span'):
                values.append(k.text)
            for j in x.find_all('a'):
                values.append(j.text)  
        values.append(soup2.find("span",attrs={"data-bind":"markupText:'currentPriceBeforePoint'"}).text)
        table.append(values)
        heads.append("Fiyat")
        head_table.append(heads)
        
```

```python
#Kullanılmak istenen kategoriler
basliklar = ["Bellek Hızı","Cihaz Ağırlığı","Dokunmatik Ekran","Ekran Boyutu","Ekran Kartı Bellek Tipi","Ekran Kartı Hafızası","Ekran Kartı Tipi","Ekran Kartı","Ekran Panel Tipi","Harddisk Kapasitesi","İşlemci Nesli","İşlemci Tipi","İşlemci","İşletim Sistemi","Kart Okuyucu","Maksimum İşlemci Hızı","Max Ekran Çözünürlüğü","Ram (Sistem Belleği)","Ram Tipi","Fiyat"]
#Kategorilere filtrelemek için istenilen kategorilerden eksik bulunan varsa tespit etmek için fonksiyon
def contains_string_array(main_array, sub_array):
    for sub_string in sub_array:
        if sub_string not in main_array:
            print(sub_string)
            return False
    return True
```

```python
for i in range(len(table)):
    print(len(table[i]),head_table[i])
```

```python
# Kullanılmak istenen özelliklere filtreleme fonksiyonu
nbyn_table=[]
for i in range(len(head_table)):
    satir=[]
    if (contains_string_array(head_table[i],basliklar)==True):
        for j in range(len(head_table[i])):
            if head_table[i][j] in basliklar:
                satir.append(table[i][j+1])#isimler eklendiği için +1 gerekli
        nbyn_table.append(satir)
        
for i in nbyn_table:
    print(i)
table=nbyn_table
```

```python
pd.DataFrame(table)
df=pd.DataFrame(table)
df.columns=basliklar
df.to_csv("dataset.csv") #.csv uzantılı olarak kaydetme 
```
